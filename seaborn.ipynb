```python
iris_df = sns.load_dataset('iris')

grid = sns.PairGrid(iris_df, hue='species', palette='Dark2')

#grid.map(plt.scatter)
grid.map_lower(plt.scatter)
grid.map_upper(sns.kdeplot)
grid.map_diag(sns.histplot)
#grid.map_offdiag()
```

```python
# custom grids

custom_grid = sns.PairGrid(
	iris_df, hue='species',
	x_vars=['sepal_length', 'sepal_width'],
	y_vars=['petal_length', 'petal_width']
)

custom_grid.map(plt.scatter)
custom_grid.add_legend()
```

```python
facet_grid = sns.FacetGrid(
	tips_df,
	col='time', row='smoker',
)
#facet_grid.add_legend()
facet_grid.map(plt.scatter, 'total_bill', 'tip')
```

```python
facet_grid = sns.FacetGrid(
	tips_df,
	col='time',	height=4, aspect=1.2, hue='smoker', col_order=['Lunch', 'Dinner'], palette='Set1'
)

facet_grid.add_legend()
facet_grid.map(sns.scatterplot, 'total_bill', 'tip', edgecolor='white')
```
